```python
def parseSheet(attrNamesPos, rowPos, sheet, posAmb):
  rowCount = 0
  attrNames = []
  ambiguous = []
  selectedAttributes = []
  for row in sheet.rows:
    if rowCount == 1:
      for i in attrNamesPos:
        attrNames.append(row[i].value)
      print(attrNames)
    if rowCount in rowPos:
      sentence = row[1].value
      amb = row[posAmb].value
      ambiguous.append(amb)
      selected = []
      colPos = 0
      for col in attrNamesPos:
        if (row[col].value is not None) and (row[col].value.lower() == 'x'):
          attrName = attrNames[colPos]
          selected.append(attrName)
        colPos += 1
      selectedAttributes.append(selected)
    rowCount += 1
  return attrNames, ambiguous, selectedAttributes

def countMetricsAttributeStrategy(golden, ambiguous, selectedAttributes):
  tpBinary, tnBinary, fpBinary, fnBinary = 0,0,0,0
  tpAttr, tnAttr, fpAttr, fnAttr = 0,0,0,0
  for gold, amb, selAttributes in zip(golden, ambiguous, selectedAttributes):
    if len(gold) == 0 and amb == 0:
      tnBinary += 1
      tnAttr += 1
    if len(gold) == 0 and amb != 0:
      fpBinary += 1
      fpAttr += len(selAttributes) ## for attribute
    if len(gold) > 0 and amb == 0:
      fnBinary += 1
      fnAttr += len(gold)
    if len(gold) > 0 and amb != 0:
      tpBinary += 1
      for attrUser in selAttributes: 
        if attrUser in gold:
          tpAttr += 1
        else:
          fpAttr += 1
      for goldAttr in gold:
        if goldAttr not in selAttributes:
          fnAttr += 1
  return tpBinary, tnBinary, fpBinary, fnBinary, tpAttr, tnAttr, fpAttr, fnAttr


def countMetricsAtLeastOne(golden, ambiguous, selectedAttributes):
  tpBinary, tnBinary, fpBinary, fnBinary = 0,0,0,0
  tpAttr, tnAttr, fpAttr, fnAttr = 0,0,0,0
  for gold, amb, selAttributes in zip(golden, ambiguous, selectedAttributes):
    if len(gold) == 0 and amb == 0:
      tnBinary += 1
      tnAttr += 1
    if len(gold) == 0 and amb != 0:
      fpBinary += 1
      fpAttr += 1
    if len(gold) > 0 and amb == 0:
      fnBinary += 1
      fnAttr += 1
    if len(gold) > 0 and amb != 0:
      tpBinary += 1
      found = False
      for attrUser in selAttributes: 
        if attrUser in gold:
          tpAttr += 1
          found = True
          break
      if not found:
        fpAttr += 1
  return tpBinary, tnBinary, fpBinary, fnBinary, tpAttr, tnAttr, fpAttr, fnAttr


def precision(tp, fp):
  return (tp / (tp + fp))

def recall(tp, fn):
  return (tp / (tp + fn))

def f1(p, r):
  return (2 * p * r)/(p + r)


def loadAnnotations(annotators, files, sheetName, attrNamesPos, rowPos, posAmb):
  annotations = {}
  for annotator in annotators:
    for fileName in files:
      if annotator in fileName:
        excelFile = BASE_PATH + fileName
        print(excelFile)
        wb = load_workbook(excelFile)
        sheet = wb[sheetName]
        attrNames, ambiguous, selectedAttributes = parseSheet(attrNamesPos, rowPos, sheet, posAmb)
        annotations[annotator] = attrNames, ambiguous, selectedAttributes
  return annotations

def cumulativeStrategy(annotations, golden, atLeast = False):
  tpBinary, tnBinary, fpBinary, fnBinary, tpAttr, tnAttr, fpAttr, fnAttr = 0, 0, 0, 0, 0, 0, 0, 0
  for annotator, annotated in annotations.items():
    ambiguous = annotated[1]
    selectedAttributes = annotated[2]
    if not atLeast:
      tpB, tnB, fpB, fnB, tpA, tnA, fpA, fnA = countMetricsAttributeStrategy(golden, ambiguous, selectedAttributes)
    else:
      tpB, tnB, fpB, fnB, tpA, tnA, fpA, fnA = countMetricsAtLeastOne(golden, ambiguous, selectedAttributes)
    tpBinary += tpB
    tnBinary += tnB
    fpBinary += fpB
    fnBinary += fnB
    tpAttr += tpA
    tnAttr += tnA
    fpAttr += fpA
    fnAttr += fnA

  pBinary = precision(tpBinary, fpBinary)
  rBinary = recall(tpBinary, fnBinary)
  f1Binary = f1(pBinary, rBinary)
  #print("Binary:", pBinary, rBinary, f1Binary)
  pAttr = precision(tpAttr, fpAttr)
  rAttr = recall(tpAttr, fnAttr)
  f1Attr = f1(pAttr, rAttr)
  #print("Attribute:", pAttr, rAttr, f1Attr)
  return pBinary, rBinary, f1Binary, pAttr, rAttr, f1Attr

def meanStrategy(annotations, golden, atLeast = False):
  pB = []
  rB = []
  f1B = []
  pA = []
  rA = []
  f1A = []
  for annotator, annotated in annotations.items():
    #print(annotator)
    ambiguous = annotated[1]
    selectedAttributes = annotated[2]
    if not atLeast:
      tpBinary, tnBinary, fpBinary, fnBinary, tpAttr, tnAttr, fpAttr, fnAttr = countMetricsAttributeStrategy(golden, ambiguous, selectedAttributes)
    else:
      tpBinary, tnBinary, fpBinary, fnBinary, tpAttr, tnAttr, fpAttr, fnAttr = countMetricsAtLeastOne(golden, ambiguous, selectedAttributes)
    pBinary = precision(tpBinary, fpBinary)
    rBinary = recall(tpBinary, fnBinary)
    f1Binary = f1(pBinary, rBinary)
    #print(pBinary, rBinary, f1Binary)
    pB.append(pBinary)
    rB.append(rBinary)
    f1B.append(f1Binary)
    pAttr = precision(tpAttr, fpAttr)
    rAttr = recall(tpAttr, fnAttr)
    f1Attr = f1(pAttr, rAttr)
    #print(pAttr, rAttr, f1Attr)
    pA.append(pAttr)
    rA.append(rAttr)
    f1A.append(f1Attr)

  pBinary = np.asarray(pB).mean()
  rBinary = np.asarray(rB).mean()
  f1Binary = np.asarray(f1B).mean()
  pAttr = np.asarray(pA).mean()
  rAttr = np.asarray(rA).mean()
  f1Attr = np.asarray(f1A).mean()
  return pBinary, rBinary, f1Binary, pAttr, rAttr, f1Attr
```
